```python
import re
def cleanResume(text):
    cleanText = re.sub('http\S+\s', ' ' ,text)
    cleanText = re.sub('@\S+', ' ' ,cleanText)
    cleanText = re.sub('#\S+', ' ' ,cleanText)
    cleanText = re.sub('[%s]' % re.escape("""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"""), ' ', cleanText)
    cleanText = re.sub(r'[^\x00-\x7f]', ' ', cleanText) 
    cleanText = re.sub('\s+', ' ', cleanText)
    #special
    cleanText = re.sub('RT|cc', ' ', cleanText)
    return cleanText
```

```python
cleanResume("hello boy #okokok httpslajajlf ok baby @gmail.com")
```

```python
df['Resume'] = df['Resume'].apply(lambda x: cleanResume(x))
```

```python
df['Resume'][0]
```

```python
#categorical values making from word
```

```python
from sklearn.preprocessing import LabelEncoder
lebel= LabelEncoder()
```

```python
lebel.fit(df['Category'])
df['Category'] = lebel.transform(df['Category'])
```

```python
#vactorization
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words = 'english')

tfidf.fit(df['Resume'])
requiredText =  tfidf.transform(df['Resume'])
```

```python
df
```

```python
#splitting
from sklearn.model_selection  import train_test_split
x_train,x_test,y_train,y_test = train_test_split(requiredText, df['Category'],test_size=0.2,random_state = 42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
#training
```

```python
from sklearn.neighbors import KNeighborsClassifier

# Example usage
clf = KNeighborsClassifier()
clf.fit(x_train, y_train)

y_predict = clf.predict(x_test)


print(y_predict)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.multiclass import OneVsRestClassifier

# Example usage
clf = OneVsRestClassifier(KNeighborsClassifier())
clf.fit(x_train, y_train)

y_predict = clf.predict(x_test)

print(y_predict)
#************************Hey WARNING**************
#Don't run all these cell when creating clf this may create conflict between clf and tfidf file.I got error doint that
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import accuracy_score
# Example usage
clf = OneVsRestClassifier(KNeighborsClassifier())
clf.fit(x_train, y_train)

y_predict = clf.predict(x_test)

print(accuracy_score(y_test,y_predict))
```
